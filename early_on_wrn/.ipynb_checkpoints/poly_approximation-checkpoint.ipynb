```python
import numpy as np
import matplotlib.pyplot as plt

def appro(a, b, beta):
    
    output = []
    for (mu, sigma) in zip(a, b):
        
        x = np.random.normal(mu, sigma*beta, 3000000)
        x_true = np.random.normal(mu, sigma, 3000000)
#         _ = plt.hist(x, 30, density=True)
#         plt.show()
        
        y = [s if s>0 else 0 for s in x]
        z = np.polyfit(x, y, 2)
        out = [l for l in z]
        output.append(out)
        
        # plot
        x1 = np.arange(np.min(x_true), np.max(x_true),0.01)
        y1 = [s if s>0 else 0 for s in x1]
        z1 = out[0]*x1*x1+out[1]*x1+out[2]
        plt.plot(x1, y1)
        plt.plot(x1, z1, 'r')
        plt.show()
        print(np.mean(np.abs(y1-z1)))
        
    print(output)
    
#     return output
```

```python
appro([-0.2, -0.28, -0.46, -0.54, -0.67, -0.59, -0.68, -0.49], 
      [0.5, 0.52, 0.66, 0.8, 0.84, 0.89, 0.94, 0.955], beta=2)
```

## one time tuning

```python
x = np.random.normal(-0.2, 0.5, 3000000) 
print(np.min(x), np.max(x))
_ = plt.hist(x, 30, density=True)
```

```python
coef = appro([-0.2], [1.5])

x1 = np.arange(np.min(x), np.max(x),0.01)
y1 = coef[0][0]*x1*x1 + coef[0][1]*x1 + coef[0][2]
plt.plot(x1, y1)
y2 = [s if s>0 else 0 for s in x1]
plt.plot(x1, y2, 'r')
```
